### 1 环境准备

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.available
plt.style.use("seaborn")

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

import warnings
warnings.filterwarnings('ignore')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

### 2 数据读取

```python
X_cityA = np.load('../../data/A/cityA/X.npy')
Y_cityA = np.load('../../data/A/cityA/Y.npy')

X_cityB = np.load('../../data/A/cityB/X.npy')
Y_cityB = np.load('../../data/A/cityB/Y.npy')

X_cityA.shape, Y_cityA.shape, X_cityB.shape, Y_cityB.shape
```

```python
X_a = X_cityA.reshape(-1, X_cityA.shape[2])
X_b = X_cityB.reshape(-1, X_cityB.shape[2])

X_a = pd.DataFrame(X_a, columns=[f'col_{num}' for num in range(X_a.shape[1])])
X_b = pd.DataFrame(X_b, columns=[f'col_{num}' for num in range(X_b.shape[1])])

Y_a = pd.DataFrame(Y_cityA)
Y_b = pd.DataFrame(Y_cityB)

X_a.shape, Y_a.shape, X_b.shape, Y_b.shape
```

### 5 探索性分析

```python
# 热力图 —— X_a
temp = X_a.copy()
temp.columns = [str(i) for i in range(temp.shape[1])]
plt.figure(figsize=(100,60))
ax = sns.heatmap(temp.corr(),xticklabels=1,yticklabels=1)
fig = ax.get_figure()

# 直方图
X_a.hist(bins=600,figsize=(40, 30))

# 趋势图
X_a.plot.line(figsize=(60, 50),subplots=True)

# 箱线图
temp = X_a.copy()
temp.columns = [str(i) for i in range(temp.shape[1])]
temp.plot(kind='box',figsize=(100, 70),subplots=True,sharey=False)
```

```python
# 热力图 —— X_b
temp = X_b.copy()
temp.columns = [str(i) for i in range(temp.shape[1])]
plt.figure(figsize=(100,60))
ax = sns.heatmap(temp.corr(),xticklabels=1,yticklabels=1)
fig = ax.get_figure()

# 直方图
X_b.hist(bins=600,figsize=(40, 30))

# 趋势图
X_b.plot.line(figsize=(60, 50),subplots=True)

# 箱线图
temp = X_b.copy()
temp.columns = [str(i) for i in range(temp.shape[1])]
temp.plot(kind='box',figsize=(100, 70),subplots=True,sharey=False)
```
